# Предварительная работа с данными

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/titanic.csv')
```

```python
data
```

```python
columns_target = ['Survived']
columns_train = ['Pclass','Sex','Age','Fare']
```

```python
X = data[columns_train]
Y = data[columns_target]
```

проверяем, есть ли пустые ячейки в колонках

```python
X['Sex'].isnull().sum()
```

```python
X['Pclass'].isnull().sum()
```

```python
X['Fare'].isnull().sum()
```

```python
X['Age'].isnull().sum()
```

заполняем пустые ячейки медианными значением по возрасту

```python
pd.options.mode.chained_assignment = None
```

```python
X['Age'] = X['Age'].fillna(X['Age'].median())
```

```python
X['Age']
```

```python
X['Age'].isnull().sum()
```

заменяем male и female на 0 и 1 с помощью словаря

```python
d = {'male':0, 'female': 1}
```

```python
X['Sex'] = X['Sex'].apply(lambda x:d[x])
```

```python
X['Sex'].head()
```

# Разделяем нашу выборку на обучающую и тестовую

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.33, random_state = 42)
```

# Загружаем модель Support Vector Machine для обучения

```python
from sklearn import svm
```

```python
premodel = svm.LinearSVC()
```

Обучаем модель с помощью нашей обучающей выборки

```python
premodel.fit(X_train, Y_train)
```

Предсказываем на тестовой выборке

```python
premodel.predict(X_test[0:10])
```

проверяем точность предсказаний

```python
premodel.score(X_test, Y_test)
```

# Применяем алгоритм дерева решений 

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
clf = tree.DecisionTreeClassifier(max_depth=5, random_state=21)
```

Обучаем

```python
clf.fit(X_train, Y_train)
```

Проверяем на тестовых данных

```python
clf.score(X_test, Y_test)
```
